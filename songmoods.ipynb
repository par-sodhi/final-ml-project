```python
import pandas as pd

df = pd.read_csv("data_moods.csv")
df.head()
```

```python
df.info()
```

There are no null values in the training set.

The target is mood. It is unlikely that `length`, `name`,`album`, `artist`, `id`, and `release_date` would be helpful in identifying the mood, so I would drop them. Also, drop the columns not in the training set.

```python
df.drop(columns=['name','album','artist','id','release_date', 'length', 'time_signature'], inplace=True)
df.info()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(df.iloc[:,:-1], df.mood, test_size=0.1)
```

```python
train_x.info()
```

Prepare the data for classification

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(train_y)
train_y = le.transform(train_y)
test_y = le.transform(test_y)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

xgb = XGBClassifier(objective="reg:squarederror", learning_rate=0.085, max_depth=11, subsample=0.65, 
                    colsample_bytree=0.845, reg_alpha=0.303, reg_lambda=0.096,gamma=0.01,n_estimators=94)
xgb.fit(train_x, train_y)
accuracy_score(test_y, xgb.predict(test_x))
```

Accuracy is good enough.

```python
data = pd.read_csv("data.csv")
data.info()
```

There are no null values in the test set either.

Make the test set columns align with the training set.

```python
data_id = data[['name', 'artists', 'id']]
data= data[['popularity', 'danceability', 'acousticness', 'energy', 'instrumentalness', 'liveness', 'valence', 'loudness', 'speechiness', 'tempo', 'key']]
```

```python
results = xgb.predict(data)
results = le.inverse_transform(results)
data_id['mood'] = results
data_id.head()
```

```python
data_id.to_csv('songmoods.csv')
```
